```python
import pandas as pd

TIME_COLUMN = "_time"
ENERGY_QTY_COLUMN = "_value"
DRIVER_COLUMN = "driver"

REDUNDANT_COLUMNS = {"Unnamed: 0", "result", "table", "_start", "_stop", "_field", "_measurement"}
EXPECTED_FINAL_COLUMNS = {TIME_COLUMN, ENERGY_QTY_COLUMN, DRIVER_COLUMN}
```

```python
df = pd.read_csv('data.csv', comment="#")
df = df.drop(columns=REDUNDANT_COLUMNS, errors='ignore')

assert set(df.columns) == EXPECTED_FINAL_COLUMNS, \
    f"Input csv structure has been changed. actual={df.columns.to_list()}"

df["_time"] = pd.to_datetime(df['_time'])
df['month'] = df['_time'].dt.month
df['year'] = df['_time'].dt.year
df
```
